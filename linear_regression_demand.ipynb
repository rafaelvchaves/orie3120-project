```python
weather = pd.read_csv('2962063.csv')
```

```python
weather.head()
```

```python
weather['Date'] = pd.to_datetime(weather['DATE']).dt.normalize()
```

```python
date = date.merge(weather, on='Date', how='inner')
date = date[['Date','hvfhs_license_num','CASE_COUNT','PRCP']]
```

```python
date
```

```python
date['day_of_week'] = date['Date'].dt.strftime('%A')
```

```python
date
```

```python
cal = calendar()
holidays = cal.holidays(start='2021-01-01', end='2021-12-31')

date['Holiday'] = date['Date'].isin(holidays)
```

```python
is_weekend = []
for i in range(len(date)):
    if date['day_of_week'].iloc[i] == 'Saturday' or date['day_of_week'].iloc[i] == 'Sunday':
        is_weekend.append(1)
    else:
        is_weekend.append(0)
date['is_weekend']=is_weekend
```

```python
date
```

```python
is_holiday = []
for i in range(len(date)):
    if date['Holiday'].iloc[i] == True:
        is_holiday.append(1)
    else:
        is_holiday.append(0)
date['is_holiday']=is_holiday
#is_holiday
```

```python
date = date.rename(columns={'hvfhs_license_num': 'Number_of_Trips', 
                            'CASE_COUNT': 'COVID_Cases',
                            'PRCP': 'Precipitation',
                            'is_weekend': 'Weekend',
                            'is_holiday': 'Holiday'})
```

```python
date
```

```python
date = date[['Date','Number of Trips','COVID Cases','Precipitation','Weekend','Holiday']]
y = date['Number of Trips']
features = date[date.columns[2:]]
#x = sm.add_constant(features)
model = sm.OLS(y,features).fit()
model.summary()
```
